```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load sample dataset for demonstration
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# Creating advanced visualizations
plt.figure(figsize=(10, 6))
```

**Creating Advanced Visualizations**
1. Box Plots:
* Box plots summarize the distribution of a continuous variable.
* Use sns.boxplot() function to create a box plot.

2. Violin Plots:
* Violin plots display the distribution of a continuous variable along with its probability density.
* Use sns.violinplot() function to create a violin plot.

3. Pair Plots:
* Pair plots visualize pairwise relationships in a dataset.
* Use sns.pairplot() function to create a pair plot.


## Leveraging Seaborn's Themes and Color Palettes
**Themes:**
* Seaborn provides built-in themes to style plots for better presentation.
* Themes include darkgrid, whitegrid, dark, white, and ticks.

**Color Palettes:**
* Seaborn offers various color palettes to customize the colors of plots.
* Palettes include deep, muted, bright, pastel, dark, and colorblind.

```python
# Box Plot
plt.subplot(1, 3, 1)
sns.boxplot(x='species', y='sepal_length', data=iris)

# Violin Plot
plt.subplot(1, 3, 2)
sns.violinplot(x='species', y='sepal_length', data=iris)
```

```python
# Violin Plot
plt.subplot(1, 3, 2)
sns.violinplot(x='species', y='sepal_length', data=iris)
```

```python
# Pair Plot
plt.subplot(1, 3, 3)
sns.pairplot(iris, hue='species')

plt.tight_layout()
plt.show()
```

```python
# Leveraging Seaborns Themes and Color Palettes
plt.figure(figsize=(10, 6))
```

```python
# Darkgrid Theme
plt.subplot(1, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=iris)
sns.set_theme(style='darkgrid')
```

```python
# Color Palette
plt.subplot(1, 2, 2)
sns.boxplot(x='species', y='sepal_length', data=iris, palette='colorblind')

plt.tight_layout()
plt.show()
```
